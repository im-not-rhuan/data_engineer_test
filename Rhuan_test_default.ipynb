```python
df = pd.read_excel('karhub_autoparts_1.xlsx')
df
```

Lendo o .xlsx como um dataframe pandas

```python
df['Valor Atributo'] = df['Valor Atributo'].str.replace(',', '.')
```

Na coluna Valor Atributo, trocamos vírgulas por pontos, para deixar os dados limpos

```python
df = df.pivot_table(index=['Nome SKU', 'Fabricante', 'Código', 'Composição', 'Categoria'], columns='Nome Atributo', values='Valor Atributo',
                     aggfunc='first')
df = df.reset_index()
df
```

```python
df_xlsx = df.fillna({'Motor': '-', 'Número de Espirais': 0})
df_xlsx
```
